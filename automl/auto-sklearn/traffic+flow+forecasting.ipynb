## Data description
- tra_X_te (1x840)    - test set input data: traffic indices for 840 contiguous quarter-hours each element is a 36*48 matrix: 36 spatial locations by 48 features
- tra_X_tr (1x1261)   - training set input data: traffic indices for 1261 contiouous quarter-hours. Each element is a 36*48 matrix: 36 spatial locations by 48 features
- tra_Y_te (36x840)   - test set output data: traffic flow for 36 locations in 840 contiguous quarter-hours from 2017-01-02 00:00
- tra_Y_tr (36x1261)  - training set output data: traffic flowfor 36 locations in 1261 contiouous quarter-hours until 2017-02-01 00:15
- tra_adj_mat (36x36) - adjacency matrix denoting the spatial connectivity of traffic network among 36 locations
- tra_X_te (1x840)    - test set input data: traffic indices for 840 contiguous quarter-hours. Each element is a 36*48 matrix: 36 spatial locations by 48 features
