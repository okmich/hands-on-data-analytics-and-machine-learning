```python
%matplotlib inline

from autoviz import AutoViz_Class
AV = AutoViz_Class()

import os

filename = os.path.join('..', '..', 'data', 'mortality.csv')
target_variable = "outcome"

dft = AV.AutoViz(
    filename,
    sep=",",
    depVar=target_variable,
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="bokeh",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```
